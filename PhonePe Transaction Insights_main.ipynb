```python
#Connect to the MySQL server
#myconnection = pymysql.connect(host='127.0.0.1',user='root',passwd='Praveen@1932000')

import mysql.connector
from mysql.connector import Error

def check_missing_tables():
    try:
        connection = mysql.connector.connect(
        host="127.0.0.1",
        user="root",
        password="Praveen@1932000",
        auth_plugin="mysql_native_password"
        )
        
        if connection.is_connected():
            cursor = connection.cursor()
            
            # List of required tables
            required_tables = [
                'aggregated_insurance',
                'aggregated_transaction',
                'aggregated_user',
                'map_insurance',
                'map_transaction',
                'map_user',
                'top_insurance',
                'top_transaction',
                'top_user'
            ]
            
            missing_tables = []
            
            # Check each table
            for table in required_tables:
                cursor.execute(f"""
                    SELECT COUNT(*)
                    FROM information_schema.tables
                    WHERE table_schema = 'phonepe_pulse' 
                    AND table_name = '{table}'
                """)
                
                if cursor.fetchone()[0] == 0:
                    missing_tables.append(table)
            
            # Print results
            if missing_tables:
                print("The following tables need to be created:")
                for table in missing_tables:
                    print(f"- {table}")
            else:
                print("All required tables already exist in the database.")
            
    except Error as e:
        print(f"Error checking tables: {e}")
    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()

if __name__ == "__main__":
    check_missing_tables()
```
